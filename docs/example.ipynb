```python
from goa import algorithms, problems
```

# Problem examples

```python
ackley_problem = problems.Ackley(bounds=(-2.5, 2.5))
ackley_problem.plot()
```

```python
quadratic_problem = problems.Quadratic(bounds=(-2.5, 2.5))
quadratic_problem.plot()
```

# Memetic Differential Evolution algorithm
### (using Coordinate Method with Simple Descent Direction)

```python
algorithms.memetic_differential_evolution(
    problem=ackley_problem,
    local_search_algorithm=algorithms.coordinate_method,
    animation_filepath='./media/01-MDE.gif'
)
```

```python
algorithms.memetic_differential_evolution(
    problem=quadratic_problem,
    local_search_algorithm=algorithms.coordinate_method,
    animation_filepath='./media/02-MDE.gif'
)
```

# Differential Evolution algorithm

```python
algorithms.differential_evolution(
    ackley_problem,
    animation_filepath='./media/01-DE.gif'
)
```

```python
algorithms.differential_evolution(
    quadratic_problem,
    animation_filepath='./media/02-DE.gif'
)
```
